### Task: Modify the code to compute the median and standard deviation instead of the mean, and apply it to a 1D array of 20 random floats between 0 and 10.

### Task: Modify the code to filter rows where age is greater than 25 and city is "New York".

```python
import pandas as pd
```

```python
data={
    'name' :['Alice','Bob','Charlie','David'],
    'age' : [25,30,35,20],
    'city': ['New York','Los Angels','New York','Chicago']
}
df=pd.DataFrame(data)
filtered_df=df[(df['age']>25) & (df['city']=='New York')]
print('Filtered Dataframe:\n',filtered_df)
```

### Task: Modify the code to group by city and compute the maximum age for each city.

```python
data2={
    'name':['Alice','Bob','Charlie','David','Eve'],
    'age':[25,30,35,20,28],
     'city':['New York','Los Angels','New York','Chicago','Los Angels']
}
df2=pd.DataFrame(data2)
grouped_df2=df2.groupby('city')['age'].max()
print('Grouped Dataframe: \n',grouped_df2)
```

### Task: Modify the code to create a boxplot of total_bill grouped by day, with points colored by sex

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
tips_plot=sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    
)
tips_plot.set(
    title='Total Bill by Day and Sex',
    xlabel='Day',
    ylabel='Total Bill'
)
plt.show()
```

### Task: Modify the code to create a heatmap of the correlation matrix for numerical columns in the tips dataset, with annotations.


```python
num=tips.select_dtypes(include=['number'])
tips_heatmap=sns.heatmap(
    num.corr(),
    annot=True,
    cmap="coolwarm"
)

tips_heatmap.set_title('Correlation Between Total Bill, Tip and Size')
plt.show()


```

### Task: Modify the code to create a scatter plot of total_bill vs. tip, with points colored by day and sized by size.


```python
import plotly.express as px
```

```python
fig=px.scatter(
    data_frame=tips,
    x='total_bill',
    y='tip',
    color='day',
    size='size',
    hover_data=['time', 'smoker'],
)
fig.update_layout(
    title='Total Bill VS Tip'
)

fig.show()
```

### Task: Modify the code to create a bar plot showing the average total_bill per day, with bars colored by sex.


```python
grouped_tips_by_day=tips.groupby(['day','sex'])['total_bill'].mean().reset_index()
plot=px.bar(
    data_frame=grouped_tips_by_day,
    x='day',
    y='total_bill',
    color='sex'
)
plot.update_layout(
    title='Average Total Bill by Day and Sex'
)
plot.show()
```
